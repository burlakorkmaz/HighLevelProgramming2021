1\. Perform the following operations on plain `txt` files:

+ create a list of integrer numbers and then save it to a text file named "data_int.txt". Run the `cat` command to print the content of the file.
+ create a matrix of 5x5 floats and then save it to a text file named "data_float.txt". Use the `cat` command to print the content of the file.
+ load the txt file of the previous point and convert it to a csv file by hand.

```python
import os
from random import randrange
int_list = []
for i in range(100):
    int_list.append(randrange(100))

with open("data_int.txt", "w") as output:
    output.write(str(int_list))
    print("data_int.txt file saved to working directory\n")
!type data_int.txt
print("\n")
#--------------------------------------------------------------
float_5x5 = np.random.rand(5,5)

with open("data_float.txt", "w") as output:
    output.write(str(float_5x5))
    print("data_float.txt file saved to working directory\n")
!type data_float.txt
print("\n")
#--------------------------------------------------------------
characters = "[]"
with open("data_float.csv", "w") as output:
    with open("data_float.txt", "r") as input_data:
        for row in input_data:
            for character in characters:
                row = row.replace(character, "")
            
            for i in range(len(row.split())):
                output.write(str(row.split()[i]) + ",")
            output.write("\n")
!type data_float.csv
```

2\. Load the file *user_data.json*, which can be found at:

and filter the data by the "CreditCardType" when it equals to "American Express". Than save the data to a new CSV file.

```python
import json
user_data = json.load(open("user_data.json"))
with open("user_data.csv", mode = 'w') as output:
    for i in range(len(user_data)):
        if user_data[i]["CreditCardType"] == "American Express":
            output.write(user_data[i]["ID"] + ',' + user_data[i]["JobTitle"] + ',' + user_data[i]["EmailAddress"] + ',' + 
            user_data[i]["FirstNameLastName"] + ',' + user_data[i]["CreditCard"] + ',' + user_data[i]["CreditCardType"] + "\n")
!type user_data.csv
```

4\. Get the binary file named *credit_card.dat* from this address:

- https://www.dropbox.com/s/8m0syw2tkul3dap/credit_card.dat

and convert the data into the real credit card number, knowing that:
- each line corresponds to a credit card number, which consists of 16 characters (which are numbers in the 0-9 range) divided in 4 blocks, with a whitespace between each block
- each character is written using a 6 bit binary representation (including the whitespace)
- the final 4 bits of each line are a padding used to determine the end of the line, and can be ignored

*Hint*: convert the binary numbers to the decimal representation first, and then use the `chr()` function to convert the latter to a char

```python
credit_card = [i.strip().split() for i in open("credit_card.dat").readlines()]

card_numbers = []
for i in range(len(credit_card)):
    card_numbers.append(credit_card[i][0][0:114])
    
card_numbers.pop(-1)

for binary in card_numbers:
    card_number = ""
    for i in range(0,len(binary),6):
        card_number = card_number + chr(int(binary[i:i+6],2))
    print(card_number)
```
